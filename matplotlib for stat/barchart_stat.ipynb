```python
import matplotlib.pyplot as plt


categories = ['Female', 'Male']
absolute_frequencies = [3, 7]
relative_frequencies = [0.3, 0.7]

# Creating bar plot for absolute frequencies
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(categories, absolute_frequencies, color='blue')
plt.xlabel('Gender')
plt.ylabel('Absolute Frequency')
plt.title('Bar Chart with Absolute Frequencies')

# Creating bar plot for relative frequencies
plt.subplot(1, 2, 2)
plt.bar(categories, relative_frequencies, color='red')
plt.xlabel('Gender')
plt.ylabel('Relative Frequency')
plt.title('Bar Chart with Relative Frequencies')

plt.tight_layout()
plt.show()
```
